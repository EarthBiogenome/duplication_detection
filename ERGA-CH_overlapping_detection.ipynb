```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the uploaded CSV file to analyze it
file_path = 'ERGA-CH sequencing status.csv'
data = pd.read_csv(file_path)

# Display the column names to identify column K
data.columns
```

```python
status_counts = data['sequencing_status_erga-ch'].value_counts()
```

```python
# Sort the values by count in descending order and create a horizontal bar chart
status_counts_sorted = status_counts.sort_values(ascending=True)

# Create a horizontal bar chart with specified x-axis intervals
plt.figure(figsize=(10, 6))
status_counts_sorted.plot(kind='barh')
plt.title('ERGA-CH Sequencing Status', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.yticks(fontsize=10)  # Keep the labels for the bars but hide the ylabel
plt.gca().set_ylabel('')  # Remove the ylabel
plt.xticks(ticks=range(0, 41, 10), fontsize=10)
plt.xlim(0, 40)
plt.tight_layout()

# Display the chart
plt.show()
```

```python
# Reorder the sequencing status as specified
desired_order = ['insdc_open', 'open', 'in_progress', 'sample_acquired'] 
status_counts_reordered = status_counts.reindex(desired_order)

# Create a horizontal bar chart with the specified order
plt.figure(figsize=(10, 6))
bars = status_counts_reordered.plot(kind='barh', color='skyblue')

plt.title('ERGA-CH Sequencing Status', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.yticks(fontsize=10)
plt.gca().set_ylabel('')
plt.xticks(ticks=range(0, 41, 10), fontsize=10)
plt.xlim(0, 40)

# Add vertical grid lines
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Annotate the counts on the right side of each bar
for index, value in enumerate(status_counts_reordered):
    if not pd.isna(value):  # Check for NaN values
        plt.text(value + 0.5, index, str(int(value)), va='center', fontsize=10)

plt.tight_layout()
plt.show()

```

```python
# Update overlap counts for the "insdc_open" group
insdc_open_total = status_counts_sorted.get("insdc_open", 0)
atlasea_overlap_count = 1

# Update the DToL overlap count
dtol_overlap_count = 5

# Update the Psyche overlap count
psyche_overlap_count = 2

# Define the color variables again
erga_ch_color = "#5DADE2"  # Soft blue for ERGA-CH
dtol_color = "#F5B041"     # Soft orange for DToL
psyche_color = "#58D68D"   # Soft green for Psyche
atlasea_color = "#AF7AC5"  # Soft purple for ATLASEA

# Initialize the figure
plt.figure(figsize=(12, 8))

# Base bar for ERGA-CH
bars = plt.barh(
    status_counts_sorted.index, 
    status_counts_sorted, 
    color=erga_ch_color, 
    edgecolor="black", 
    label="ERGA-CH"
)

# Overlay bars for overlaps
if "in_progress" in status_counts_sorted.index:
    in_progress_index = list(status_counts_sorted.index).index("in_progress")
    # DToL overlap
    plt.barh(
        status_counts_sorted.index[in_progress_index], 
        dtol_overlap_count, 
        color=dtol_color, 
        edgecolor="black", 
        label="DToL Overlap"
    )
    # Psyche overlap
    plt.barh(
        status_counts_sorted.index[in_progress_index], 
        psyche_overlap_count, 
        left=dtol_overlap_count, 
        color=psyche_color, 
        edgecolor="black", 
        label="Psyche Overlap"
    )

if "insdc_open" in status_counts_sorted.index:
    insdc_open_index = list(status_counts_sorted.index).index("insdc_open")
    # ATLASEA overlap
    plt.barh(
        status_counts_sorted.index[insdc_open_index], 
        atlasea_overlap_count, 
        color=atlasea_color,  
        edgecolor="black", 
        label="ATLASEA Overlap"
    )

# Add counts as text to each bar
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.2, bar.get_y() + bar.get_height() / 2, str(int(width)), va="center", fontsize=10)

if "in_progress" in status_counts_sorted.index:
    # Add count for DToL overlap
    plt.text(
        dtol_overlap_count - 1, 
        bars[in_progress_index].get_y() + bars[in_progress_index].get_height() / 2, 
        str(dtol_overlap_count), 
        va="center", 
        ha="right", 
        fontsize=10, 
        color="white"
    )
    # Add count for Psyche overlap
    plt.text(
        dtol_overlap_count + psyche_overlap_count - 1, 
        bars[in_progress_index].get_y() + bars[in_progress_index].get_height() / 2, 
        str(psyche_overlap_count), 
        va="center", 
        ha="right", 
        fontsize=10, 
        color="white"
    )

if "insdc_open" in status_counts_sorted.index:
    # Add count for ATLASEA overlap
    plt.text(
        atlasea_overlap_count - 0.5, 
        bars[insdc_open_index].get_y() + bars[insdc_open_index].get_height() / 2, 
        str(atlasea_overlap_count), 
        va="center", 
        ha="right", 
        fontsize=10, 
        color="white"
    )

# Adjustments for axis labels and title
plt.title("ERGA-CH Sequencing Status with Project Overlaps", fontsize=14)
plt.xlabel("Count", fontsize=12)
plt.ylabel("Sequencing Status", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(range(len(status_counts_sorted)), status_counts_sorted.index, fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.3) # Add grid lines for x-axis

# Add legend and tight layout
plt.legend(loc="lower right", fontsize=14)
plt.tight_layout()

# Display the chart
plt.show()
```
